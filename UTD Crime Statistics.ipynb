```python
duidwi = 0

#minors in possession or consumption of alcohol
#public intoxication
alcohol=0

#other driving crimes. Invalid License, 
other_driving=0

marijuana=0
drug=0

theft = 0
burglary = 0
assault = 0
harassment=0

#terroristic threats. Real concerns
terror=0


graffiti=0
trespass=0
mischeif=0
suicide=0
mental=0
other_agency=0
credit_card=0
false_alarm=0

arrest = 0
assigned =0
inactivation =0


with open('crimes.csv', newline = "") as crimes_file:
    csv_reader = csv.reader(crimes_file, delimiter = ',')

    for i in csv_reader:
        one_string = '' #consolidates a single row into one string
        for x in i:  
            one_string += x
            
        if 'ARREST' in one_string:
            arrest+=1
        if 'ASSIGNED' in one_string:
            assigned+=1
        if 'INACT' in one_string:
            inactivation+=1
                
        if 'WHILE INTOXICATED' in one_string or 'UNDER INFLUENCE' in one_string:
            duidwi+=1
        elif 'ABC' in one_string or 'INTOX' in one_string:
            alcohol+=1
        elif 'DRIVING' in one_string or 'HIT AND' in one_string or 'LICENSE' in one_string or 'ACCIDENT' in one_string or 'VEHICLE' in one_string:
            other_driving+=1
        elif 'MARI' in one_string:
            marijuana+=1
        elif 'DRUG' in one_string:
            drug+=1
            
        elif 'THEFT' in one_string:
            theft+=1
        elif 'BURG' in one_string:
            burglary+=1
        elif 'ASSAULT' in one_string:
            assault+=1
        elif 'HARASSMENT' in one_string:
            harassment+=1
        elif 'TERRORISTIC THREAT' in one_string:
            terror+=1
        
        elif 'GRAFFITI' in one_string:
            graffiti+=1
        elif 'TRESPASS' in one_string:
            trespass+=1
        elif 'MISCHIEF' in one_string:
            mischeif+=1
        elif 'SUICIDE' in one_string:
            suicide+=1
        elif 'MENTAL' in one_string:
            mental+=1
        elif 'Other Agency' in one_string or 'warrant' in one_string:
            other_agency+=1
        elif 'CREDIT CARD' in one_string:
            credit_card+=1
        elif 'FALSE ALARM' in one_string or 'FIRE ALARM' in one_string:
            false_alarm+=1
                
```

```python
import re
from collections import Counter
    
with open('crimes.csv', newline = "") as crimes_file:
    csv_reader = csv.reader(crimes_file, delimiter = ',')

    words = []
    locations = []

    #for every row in the csv...
    for i in csv_reader:
        #for every item in the row, find all words.
        #indexed for loop
        for idx, x in enumerate(i):
            addwords = re.findall(r'\w+', x)
            words += addwords
            
            #the second row is the row with the locations
            if(idx == 1):
                locations.append(x)
                
    #print(locations)
    locations_counts = Counter(locations)
        
    #List but capitalized
    cap_words = [word.upper() for word in words]

    word_counts = Counter(cap_words)
    
ignore = ['THE','A','IF','IN','IT','OF','ON','AND','WAS','STUDENT','UTD','PC','PKWY']
for word in ignore:
    if word in word_counts:
        del word_counts[word]
```
